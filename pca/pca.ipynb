```python
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
df.describe()
```

```python
import scipy.stats as stats
```

```python
cols = ['Age',	'RestingBP',	'Cholesterol',	'FastingBS',	'MaxHR',	'Oldpeak',	'HeartDisease']

for col in cols:
    upper_limit = df[col].mean() + 3*df[col].std()
    lower_limit = df[col].mean() - 3*df[col].std()
    df = df[(df[col]<upper_limit) & (df[col]>lower_limit)]

df
```

```python
dummies = pd.get_dummies(df[['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope']], dtype=int)
```

```python
dummies
```

```python
df = df.drop(['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'], axis='columns')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df = pd.concat([df,dummies], axis='columns')
df
```

```python
df = scaler.fit_transform(df)
df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
pca.fit(df)
df
```
