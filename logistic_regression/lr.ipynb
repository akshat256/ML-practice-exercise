```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('HR_comma_sep.csv')
df.describe()
```

```python
df = df.drop(['average_montly_hours','time_spend_company'], axis='columns')
```

```python
for i in range(14999):
    if(df['salary'][i] == 'low'): df['salary'] = 0
    if(df['salary'][i] == 'medium'): df['salary'] = 0.5
    if(df['salary'][i] == 'high'): df['salary'] = 1
```

```python
dummies = pd.get_dummies(df['Department'])
dummies
```

```python
df = df.drop('Department', axis='columns')
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
lr = LogisticRegression()
```

```python
X = merged.drop('left',axis='columns')
y = merged['left'].values
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8)
```

```python
lr.fit(X_train,y_train)
```

```python
y_predicted = lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```
